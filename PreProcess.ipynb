```python
import pandas as pd
import numpy as np
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
df = pd.read_csv('./processed.cleveland.data', header=None, names=column_names)

```

```python
df.replace('?', np.nan, inplace=True)

# Convert columns to numeric (as '?' was originally string)
df = df.apply(pd.to_numeric, errors='ignore')

# Replace missing values (NaN) with the mean of each column
df.fillna(df.mean(), inplace=True)

# Convert the mean of each column to integers
df = df.astype(int)
```

```python
df.isnull().sum()
```

```python
df.loc[df['target'] > 1, 'target'] = 1
df.to_csv('processed_data.csv', index=False)
```
